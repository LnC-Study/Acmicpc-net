```python
def data_in():
    width, height = map(int, input().split())
    ant_x, ant_y = map(int, input().split())
    T = int(input())
    return width, height, ant_x, ant_y, T

def solve(_width, _height, _ant_x, _ant_y, _T):
    direction = 0
    while True:
        
        # right up
        if direction == 0:
            if _width - _ant_x < _height - _ant_y:
                _next = _width - _ant_x
                direction = 1
            else:
                _next = _height - _ant_y
                direction = 3
            if _next < _T:
                _ant_x = _ant_x + _next
                _ant_y = _ant_y + _next
                _T = _T - _next
                continue
            else:
                _ant_x = _ant_x + _T
                _ant_y = _ant_y + _T
                break
            
        # left up
        elif direction == 1:
            if _ant_x - 0 < _height - _ant_y:
                _next = _ant_x - 0
                direction = 0
            else:
                _next = _height - _ant_y
                direction = 2
            if _next < _T:
                _ant_x = _ant_x - _next
                _ant_y = _ant_y + _next
                _T = _T - _next
                continue
            else:
                _ant_x = _ant_x - _T
                _ant_y = _ant_y + _T
                break
                        
       # left down
        elif direction == 2:
            if _ant_x - 0 < _ant_y - 0:
                _next = _ant_x - 0
                direction = 3
            else:
                _next = _ant_y - 0
                direction = 1
            if _next < _T:
                _ant_x = _ant_x - _next
                _ant_y = _ant_y - _next
                _T = _T - _next
                continue
            else:
                _ant_x = _ant_x - _T
                _ant_y = _ant_y - _T
                break
        
        # right down
        elif direction == 3:  # right down
            if _height - _ant_x < _ant_y - 0:
                _next = _height - _ant_x
                direction = 2
            else:
                _next = _ant_y - 0
                direction = 0
            if _next < _T:
                _ant_x = _ant_x + _next
                _ant_y = _ant_y - _next
                _T = _T - _next
                continue
            else:
                _ant_x = _ant_x + _T
                _ant_y = _ant_y - _T
                break
                
    return _ant_x, _ant_y

if __name__ == '__main__':
    width, height, ant_x, ant_y, T = data_in()
    result_x, result_y = solve(width, height, ant_x, ant_y, T)
    print(result_x, result_y)
```
