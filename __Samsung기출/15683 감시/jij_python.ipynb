```python
class cam:
    def __init__(self, _row, _col, _num):
        self.row, self.col, self.num = _row, _col, _num
        
def data_in():
    N, M = map(int, input().split())
    room = list(list(map(int, input().split())) for i in range(N))
    allCamera = []
    theNumOfVacancies = 0
    for row in range(N):
        for col in range(M):
            if room[row][col] == 0:
                theNumOfVacancies = theNumOfVacancies + 1
                continue
            elif room[row][col] in [1, 2, 3, 4, 5]:
                allCamera.append(cam(row, col, room[row][col]))
    return N, M, room, allCamera, theNumOfVacancies

def cctv(_c, _N, _M, _room):
    up = 0
    down = 0
    left = 0
    right = 0
    # up
    for i in range(_c.row - 1, -1, -1):
        if _room[i][_c.col] in [1, 2, 3, 4, 5, 7]:
            continue
        elif _room[i][_c.col] == 0:
            up = up + 1
        # 7: 감시된 영역
        elif _room[i][_c.col] == 6:
            break
    # down
    for i in range(_c.row + 1, _N, 1):
        if _room[i][_c.col] in [1, 2, 3, 4, 5, 7]:
            continue
        elif _room[i][_c.col] == 0:
            down = down + 1
        elif _room[i][_c.col] == 6:
            break
    # left
    for i in range(_c.col - 1, -1, -1):
        if _room[_c.row][i] in [1, 2, 3, 4, 5, 7]:
            continue
        elif _room[_c.row][i] == 0:
            left = left + 1
        elif _room[_c.row][i] == 6:
            break
    # right
    for i in range(_c.col + 1, _M, 1):
        if _room[_c.row][i] in [1, 2, 3, 4, 5, 7]:
            continue
        elif _room[_c.row][i] == 0:
            right = right + 1
        elif _room[_c.row][i] == 6:
            break
    return up, down, left, right

def watch(_c, _room, direction, _N, _M):
    # up
    if direction == 1:
        for i in range(_c.row - 1, -1, -1):
            if _room[i][_c.col] not in [1, 2, 3, 4, 5, 6]:
                _room[i][_c.col] = 7
    # down
    elif direction == 2:
        for i in range(_c.row + 1, _N, 1):
            if _room[i][_c.col] not in [1, 2, 3, 4, 5, 6]:
                _room[i][_c.col] = 7
    # left
    elif direction == 3:
        for i in range(_c.col - 1, -1, -1):
            if _room[_c.row][i] not in [1, 2, 3, 4, 5, 6]:
                _room[_c.row][i] = 7
    # right
    elif direction == 4:
        for i in range(_c.col + 1, _M, 1):
            if _room[_c.row][i] not in [1, 2, 3, 4, 5, 6]:
                _room[_c.row][i] = 7
    return _room

if __name__ == '__main__':
    N, M, room, allCamera, theNumOfVacancies = data_in()
    allCamera.sort(reverse = True, key = lambda object: object.num)
    
    for c in allCamera:
        up, down, left, right = cctv(c, N, M, room)
        if c.num == 1:
            look = max([up, down, left, right])
            if look == 0:
                continue
            if theNumOfVacancies - look <= 0:
                theNumOfVacancies = 0
                break
            theNumOfVacancies = theNumOfVacancies - look
            if look == up:
                room = watch(c, room, 1, N, M)
            elif look == down:
                room = watch(c, room, 2, N, M)
            elif look == left:
                room = watch(c, room, 3, N, M)
            elif look == right:
                room = watch(c, room, 4, N, M)
        elif c.num == 2:
            look = max([up + down, left + right])
            if look == 0:
                continue
            if theNumOfVacancies - look <= 0:
                theNumOfVacancies = 0
                break
            theNumOfVacancies = theNumOfVacancies - look
            if look == up + down:
                room = watch(c, room, 1, N, M)
                room = watch(c, room, 2, N, M)
            elif look == left + right:
                room = watch(c, room, 3, N, M)
                room = watch(c, room, 4, N, M)
        elif c.num == 3:
            look = max([up + right, right + down, down + left, left + up])
            if look == 0:
                continue
            if theNumOfVacancies - look <= 0:
                theNumOfVacancies = 0
                break
            theNumOfVacancies = theNumOfVacancies - look
            if look == up + right:
                room = watch(c, room, 1, N, M)
                room = watch(c, room, 4, N, M)
            elif look == right + down:
                room = watch(c, room, 4, N, M)
                room = watch(c, room, 2, N, M)
            elif look == down + left:
                room = watch(c, room, 2, N, M)
                room = watch(c, room, 3, N, M)
            elif look == left + up:
                room = watch(c, room, 3, N, M)
                room = watch(c, room, 1, N, M)
        elif c.num == 4:
            look = max([up + right + left, up + right + down, down + left + right, down + left + up])
            if look == 0:
                continue
            if theNumOfVacancies - look <= 0:
                theNumOfVacancies = 0
                break
            theNumOfVacancies = theNumOfVacancies - look
            if look == up + right + left:
                room = watch(c, room, 1, N, M)
                room = watch(c, room, 4, N, M)
                room = watch(c, room, 3, N, M)
            elif look == up + right + down:
                room = watch(c, room, 1, N, M)
                room = watch(c, room, 4, N, M)
                room = watch(c, room, 2, N, M)
            elif look == down + left + right:
                room = watch(c, room, 2, N, M)
                room = watch(c, room, 3, N, M)
                room = watch(c, room, 4, N, M)
            elif look == down + left + up:
                room = watch(c, room, 2, N, M)
                room = watch(c, room, 3, N, M)
                room = watch(c, room, 1, N, M)
        elif c.num == 5:
            look = up + down + left + right
            if look == 0:
                continue
            if theNumOfVacancies - look <= 0:
                theNumOfVacancies = 0
                break
            theNumOfVacancies = theNumOfVacancies - look
            room = watch(c, room, 1, N, M)
            room = watch(c, room, 2, N, M)
            room = watch(c, room, 3, N, M)
            room = watch(c, room, 4, N, M)
    print(theNumOfVacancies)
```
